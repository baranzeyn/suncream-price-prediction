```python
import csv
import pandas as pd
import re
import numpy as np
```

```python
#Etiketleyeceğim veya temizleyeceğim veriler
tinted=["renkli","tinted","bb","pembe","beyazlaştırıcı","bright","eşitleyici","tone","light","medium,bronzlaştırıcı","bronze","beyaz"]#0
untinted=["transparant","transparent","şeffaf","renksiz"]#-1
sun_milk=["güneş sütü","sunmilk","sun milk","güneşsütü",]#1
sun_cream=["güneş kremi","sun cream","suncream","güneşkremi","güneş losyonu","güneşlosyonu","sun lotion","sunlotion","güneş koruyucu"]#2
sun_stick=["stick"]#4
sun_spray=["güneş sprey","sprey","spray","sun spray"]#3
spf_50=["spf50","50spf","50 spf","spf 50"]
spf_40=["spf40","40spf","40 spf","spf 40","spf45","45spf","45 spf","spf 45"]
spf_30=["spf30","30spf","30 spf","spf 30"]
spf_100=["spf100","100spf","100 spf","spf 100","spf90","90spf","90 spf","spf 90"]
sun=tinted+untinted+sun_cream+sun_milk+sun_stick+sun_spray+spf_100+spf_30+spf_40+spf_50
```

```python
data=r"C:\Users\baran\OneDrive\Masaüstü\Zeynep Baran\Veri Toplama\SUN CREAM-2.xlsx"
df=pd.read_excel(data)

sun_cream_types=[]
color_type=[]
product_quantity=[]
brand_names_list = []
spf_list = []
price_list = []

```

```python
for index, row in df.iterrows():
    brand_names = row["BRAND"]
    prices = row["PRICE"]
    titles = row["TITLE"]

    # Marka adı temizleme
    brand_names = brand_names.lower().replace(" ", "").replace("'", "").replace("%", "").replace("-", "")

    # Fiyat bilgisi temizleme
    prices = prices.lower().replace("tl", "").replace("t l", "").replace(" ", "").replace(".", "").replace(",", ".")
    price_float = float(prices)
    price_int = int(price_float)
    prices = prices.replace(",", "")
    
    # Başlığı temizleme ve işlemler
    #titles = titles.replace('+', '').replace('-', '').replace(',', '')
    


    if "ml" in titles or "gr" in titles:
        matches = re.finditer(r'(\d+)\s*(ml|gr)', titles.lower())
        for match in matches:
            numerical_value, unit = match.groups()
            if numerical_value.isdigit():
                print(f"Extracted Numerical Value: {numerical_value}")
                print(f"Unit: {unit}")

                
                # Sayı kontrolü ekleyerek "ml" veya "gr" ifadesinden önce sayı var mı kontrol ediyoruz
                price_list.append(price_int)

                # Ürün tipi etiketlemesi
                if any(keyword in titles for keyword in sun_milk):
                    sun_cream_types.append("güneş sütü")  # güneş sütüyse 
                elif any(keyword in titles for keyword in sun_spray):
                    sun_cream_types.append("sprey")  # spreyse 
                elif any(keyword in titles for keyword in sun_stick):
                    sun_cream_types.append("stick")  # stick 
                elif any(keyword in titles for keyword in sun_cream):
                    sun_cream_types.append("krem")  # krem 

                # Renk değeri etiketlemesi
                if any(keyword in titles for keyword in tinted):
                    color_type.append("renkli")  
                else:
                    color_type.append("renksiz")  

                # SPF değer etiketlemesi
                if any(keyword in titles for keyword in spf_50):
                    spf_list.append("50")
                elif any(keyword in titles for keyword in spf_40):
                    spf_list.append("40")
                elif any(keyword in titles for keyword in spf_30):
                    spf_list.append("30")
                elif any(keyword in titles for keyword in spf_100):
                    spf_list.append("100")
                else:
                    spf_list.append(np.nan)
                product_quantity.append(numerical_value)
                brand_names_list.append(brand_names)
```

```python
# Yeni DataFrame oluştur
df_updated = pd.DataFrame({
    "BRAND": brand_names_list,
    "SPF": spf_list,
    "TYPE": sun_cream_types,
    "COLOR": color_type,
    "PRODUCT QUANTITY": product_quantity,
    "PRICE": price_list
})

df_updated.to_excel("SUN CREAM-3.xlsx", index=False)
```
